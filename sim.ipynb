```python
from totoro import Simulator
sim = Simulator("Greedy")
sim.load_sim("testset/test/003.txt", packet_num=1000)
sim.reset()
sim.simulate()
```

```python
import numpy as np
# import scipy.stats as stats

# 參數設定
alpha_1, beta_1 = 5, 7  # p1 的貝塔分佈參數
alpha_2, beta_2 = 3, 5  # p2 的貝塔分佈參數

# 蒙特卡洛模擬
n_samples = 100000
p1_samples = np.random.beta(alpha_1, beta_1, n_samples)
p2_samples = np.random.beta(alpha_2, beta_2, n_samples)

# 計算 p = p1 * p2
p_samples = p1_samples * p2_samples

# 計算 95% 信賴區間
p_ci_lower, p_ci_upper = np.percentile(p_samples, [2.5, 97.5])

p_ci_lower, p_ci_upper
```

```python
import scipy.stats as stats

# Beta distribution parameters
alpha = 6  # 1 + k
beta = 6   # 1 + (n - k)

# Calculate the 95% credible interval
credible_interval = stats.beta.ppf([0.025, 0.975], alpha, beta)
credible_interval
```
